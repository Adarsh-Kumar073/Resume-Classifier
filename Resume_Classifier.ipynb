```python
max_size = data['Category'].value_counts().max()
print(max_size)
```

```python
# Perform oversampling
balanced_df = data.groupby('Category').apply(lambda x: x.sample(max_size, replace=True)).reset_index(drop=True)

# Shuffle the dataset to avoid any order bias
data = balanced_df.sample(frac=1).reset_index(drop=True)
```

```python
data.Category.value_counts()
```

```python
import re
def clean_text(txt):
    clean_txt=re.sub("http:\S+\s","", txt)
    cleanText = re.sub('RT|cc', ' ', clean_txt)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
clean_text("my #### $ # I only believe in Jassi bhai, please follow @mohammed_sirajofficial on http:/instagram")
```

```python
data['Resume']= data['Resume'].apply(lambda x: clean_text(x))
```

```python
from sklearn.preprocessing import LabelEncoder
lebl = LabelEncoder()
```

```python
lebl.fit(data['Category'])
```

```python
lebl.inverse_transform([6])
```

```python
data['Category'] = lebl.transform(data['Category'])
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf= TfidfVectorizer(stop_words= 'english')
```

```python
tfidf.fit(data['Resume'])
resume  = tfidf.transform(data['Resume'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,  y_test= train_test_split(resume, data['Category'], test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
svm_model= SVC()
```

```python
svm_model.fit(x_train,y_train)
```

```python
y1_pded=svm_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y1_pded)
```

```python
confusion_matrix(y_test,y1_pded)
```

```python
# Function to predict the category of a resume
def pred(input_resume):
    # Preprocess the input text (e.g., cleaning, etc.)
    cleaned_text = clean_text(input_resume) 

    # Vectorize the cleaned text using the same TF-IDF vectorizer used during training
    vectorized_text = tfidf.transform([cleaned_text])
    
    # Convert sparse matrix to dense
    vectorized_text = vectorized_text.toarray()

    # Prediction
    predicted_category = svm_model.predict(vectorized_text)

    # get name of predicted category
    predicted_category_name = lebl.inverse_transform(predicted_category)

    return predicted_category_name[0]  # Return the category name
```
